# Parte 1: Uso de numpy

### 1. ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
arr_bool = np.zeros(1000, dtype=bool)
print(f"El tamaño en bytes es: {arr_bool.nbytes} bytes")
```

### 2. Cree un array cuyos elementos sean los enteros pares en [1,100] y en orden decreciente. Muestre los 10 últimos por pantalla.

```python
pares = np.arange(2, 101, 2)  # Genera pares del 2 al 100
pares_decrecientes = pares[::-1]  # Invierte el orden
print("Los 10 últimos elementos son:")
print(pares_decrecientes[-10:])
```

### 3. Dado el array, seleccionar con una instrucción el subarray de elementos y despues seleccionar el subarray de elementos

```python
arr = np.array([
    [2, 4, 5, 6],
    [0, 3, 7, 4],
    [8, 8, 5, 2],
    [1, 5, 6, 1]
])
print("Array original:")
print(arr)
```

```python
print("\nPrimer subarray (0 3 7 4):")
subarray1 = arr[1, :]  # Selecciona la segunda fila completa
print(subarray1)
```

```python
print("\nSegundo subarray (2 5 / 8 5):")
subarray2 = arr[0:2, 0:2]  # Selecciona las primeras 2 filas y primeras 2 columnas
print(subarray2)
```

### 4. Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])
diferencia = a - b
media_diferencia = np.mean(diferencia)
print(f"Array a: {a}")
print(f"Array b: {b}")
print(f"Diferencia (a-b): {diferencia}")
print(f"Media aritmética de la diferencia: {media_diferencia}")
```

### 5. Se tienen valores en cordenadas (x,y) en las columnas de array, agregar a el array indicado las nuevas cordenadas, mostrando en pantalla las dimensiones del nuevo array

```python
coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])
print("Coordenadas originales:")
print(coordenadas)
```

```python
nueva_coordenada = np.array([[37.1, -3.6]])
coordenadas_nuevas = np.vstack((coordenadas, nueva_coordenada))
print("\nCoordenadas con el nuevo punto agregado:")
print(coordenadas_nuevas)
print(f"Dimensiones del nuevo array: {coordenadas_nuevas.shape}")
```

### 6. De ejercicio anterior, trasponer el array y ahora agregar dos nuevos pares de corrdenadas

```python
# Copiar el array
coordenadas_copia = coordenadas_nuevas.copy()
print("Array copiado:")
print(coordenadas_copia)
```

```python
# Transponer el array
coordenadas_traspuestas = coordenadas_copia.T
print("\nArray transpuesto:")
print(coordenadas_traspuestas)
```

```python
nuevas_coordenadas = np.array([[10.8, 35.8], [3.0, 12.0]])
coordenadas_final = np.hstack((coordenadas_traspuestas, nuevas_coordenadas))
print("\nArray final con las dos nuevas coordenadas:")
print(coordenadas_final)
print(f"Dimensiones del array final: {coordenadas_final.shape}")
```

# Parte 2: Exploración y minería de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from math import sqrt
df = pd.read_csv('NucleosPoblacion.csv')
df.head()
```

### 1. ¿Cuántos Municipios tienen más de 100000 habitantes?.

```python
municipios_grandes = df[df['Poblacion'] > 100000]
cantidad_municipios_grandes = len(municipios_grandes)
print(f"Hay {cantidad_municipios_grandes} municipios con más de 100,000 habitantes.")
```

### 2. Realizar una gráfica de barras sobre la probación de cada ciudad, ondenandolar de menor a mayor y responder lo siguiente:
* Cual es la segunda ciudad mas poblada?
* Qué posición ocupa granada en el rankin de la ciudad mas poblada?

```python
# Ordenar el DataFrame por población (de menor a mayor)
df_ordenado = df.sort_values(by='Poblacion')

# Crear la gráfica
plt.figure(figsize=(12, 6))
plt.bar(df_ordenado['Municipio'], df_ordenado['Poblacion'])
plt.xticks(rotation=90)
plt.xlabel('Municipios')
plt.ylabel('Población')
plt.title('Población por Municipio (Menor a Mayor)')
plt.tight_layout()
plt.show()

# 2.a. ¿Cuál es la segunda ciudad más poblada?
df_ordenado_descendente = df.sort_values(by='Poblacion', ascending=False)
segunda_ciudad = df_ordenado_descendente.iloc[1]['Municipio'] if len(df_ordenado_descendente) > 1 else "No hay suficientes datos"
segunda_poblacion = df_ordenado_descendente.iloc[1]['Poblacion'] if len(df_ordenado_descendente) > 1 else 0
print(f"a. La segunda ciudad más poblada es {segunda_ciudad} con {segunda_poblacion} habitantes.")

if 'Granada' in df_ordenado_descendente['Municipio'].values:
    posicion_granada = df_ordenado_descendente.reset_index().index[df_ordenado_descendente.reset_index()['Municipio'] == 'Granada'][0] + 1
    print(f"Granada ocupa la posición {posicion_granada} en el ranking de ciudades más pobladas.")
else:
    print("Granada no se encuentra en el dataset.")
```

### 3. Cuantos municipios de la Rioja tienen mas de 20000 habitantes?

```python
municipios_rioja = df[(df['Provincia'] == 'La Rioja') & (df['Poblacion'] > 20000)]
cantidad_municipios_rioja = len(municipios_rioja)
print(f"Hay {cantidad_municipios_rioja} municipios en La Rioja con más de 20,000 habitantes.")
```

### 4. Cual es el municipio situado mas al norte? 

```python
municipio_norte = df.loc[df['Y'].idxmax()]
print(f"El municipio más al Norte es {municipio_norte['Municipio']} de la provincia de {municipio_norte['Provincia']}")
print(f"con una población de {municipio_norte['Poblacion']} habitantes.")
print(f"Sus coordenadas son: X={municipio_norte['X']}, Y={municipio_norte['Y']}")
```

### 5. Obtenga la media, mediana, desviación estándar y valores maximos y minimos de la población de los municipios de la provincia de granada

```python
municipios_granada = df[df['Provincia'] == 'Granada']
if len(municipios_granada) > 0:
    stats_granada = {
        'media': municipios_granada['Poblacion'].mean(),
        'mediana': municipios_granada['Poblacion'].median(),
        'desviacion_estandar': municipios_granada['Poblacion'].std(),
        'maximo': municipios_granada['Poblacion'].max(),
        'minimo': municipios_granada['Poblacion'].min()
    }
     
    print("Estadísticas de población de los municipios de Granada:")
    print(f"Media: {stats_granada['media']}")
    print(f"Mediana: {stats_granada['mediana']}")
    print(f"Desviación estándar: {stats_granada['desviacion_estandar']}")
    print(f"Valor máximo: {stats_granada['maximo']}")
    print(f"Valor mínimo: {stats_granada['minimo']}")
else:
    print("No hay municipios de Granada en el dataset de ejemplo.")
```

### 6. Ralizar histograma con la población de los municipios para cada una de las provincias 

```python
provincias = df['Provincia'].unique()
print(f"Provincias en el dataset: {provincias}")
plt.figure(figsize=(15, 10))
for i, provincia in enumerate(provincias, 1):
    plt.subplot(len(provincias)//3 + 1, 3, i)
    df_provincia = df[df['Provincia'] == provincia]
    plt.hist(df_provincia['Poblacion'], bins=10)
    plt.title(provincia)
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

### 7. Seleccionar al azar 50 municipios diferente de entre los disponibles en el archivo. Asegurar de que no se repitan. Luego calcular el promedio de la población y std.

```python
if len(df) >= 50:
    indices_aleatorios = random.sample(range(len(df)), 50)
    municipios_aleatorios = df.iloc[indices_aleatorios]
    
    promedio_poblacion = municipios_aleatorios['Poblacion'].mean()
    desviacion_estandar = municipios_aleatorios['Poblacion'].std()
     
    print(f"Promedio de población de los 50 municipios aleatorios: {promedio_poblacion}")
    print(f"Desviación estándar de la población: {desviacion_estandar}")
else:
    print(f"El dataset no tiene suficientes municipios para seleccionar 50 (tiene {len(df)}).")
```

### 8. Indicar los nombres de los municiíos mas cercanos y mas lejanos a Madrid

```python
madrid_df = df[df['Municipio'] == 'Madrid']

if len(madrid_df) > 0:
    madrid_x = madrid_df.iloc[0]['X']
    madrid_y = madrid_df.iloc[0]['Y']
    
    # Filtrar todos los municipios excepto Madrid
    df_sin_madrid = df[df['Municipio'] != 'Madrid']
    
    # Calcular la distancia euclidiana para cada municipio
    df_sin_madrid['Distancia_a_Madrid'] = df_sin_madrid.apply(
        lambda row: sqrt((row['X'] - madrid_x)**2 + (row['Y'] - madrid_y)**2), 
        axis=1
    )
    
    # Encontrar el más cercano y el más lejano
    municipio_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]
    municipio_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]
    
    print(f"El municipio más cercano a Madrid es {municipio_cercano['Municipio']}")
    print(f"con una distancia de {municipio_cercano['Distancia_a_Madrid']:.4f} unidades.")
    print()
    print(f"El municipio más lejano a Madrid es {municipio_lejano['Municipio']}")
    print(f"con una distancia de {municipio_lejano['Distancia_a_Madrid']:.4f} unidades.")
else:
    print("Madrid no se encuentra en el dataset.")
```
